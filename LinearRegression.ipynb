```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import statsmodels.api as sm
from statsmodels.api import OLS

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score

import warnings
warnings.filterwarnings("ignore")

%matplotlib inline
```

```python
# read in the data
basketball_df = pd.read_csv('data/basketball.csv')
```

#### Attribute Information:

1. height: This refers to the height in feet    
2. weight: This refers to the weight in pounds    
3. success_field_goals: This refers to the percentage of successful field goals (out of 100 that were attempted)    
4. success_free_throws: This refers tot the percentage of successful free throws (out of 100 that were attempted)    
5. avg_points_scored: This refers to the average points scored per game

```python
# descriptive statistics
basketball_df.describe()
```

```python
height = basketball_df.height
weight = basketball_df.weight
success_field_goals = basketball_df.success_field_goals
success_free_throws = basketball_df.success_free_throws
avg_points_scored = basketball_df.avg_points_scored

fig, ax = plt.subplots(2,2, figsize=(14,14))

ax[0,0].plot(height, avg_points_scored, 'o')
ax[0,0].set_xlabel('Height', fontsize=14)
ax[0,0].set_ylabel('Avg points scored', fontsize=14)
ax[0,0].set_title('Height vs Avg points scored', fontsize=14)

ax[0,1].plot(weight, avg_points_scored, 'o')
ax[0,1].set_xlabel('Weight', fontsize=14)
ax[0,1].set_ylabel('Avg points scored', fontsize=14)
ax[0,1].set_title('Weight vs Avg points scored', fontsize=14)

ax[1,0].plot(success_field_goals, avg_points_scored, 'o')
ax[1,0].set_xlabel('Successful field goals', fontsize=14)
ax[1,0].set_ylabel('Avg points scored', fontsize=14)
ax[1,0].set_title('Successful field goals vs Avg points scored', fontsize=14)

ax[1,1].plot(success_free_throws, avg_points_scored, 'o')
ax[1,1].set_xlabel('Successful free throws', fontsize=14)
ax[1,1].set_ylabel('Avg points scored', fontsize=14)
ax[1,1].set_title('Successful free throws vs Avg points scored', fontsize=14)
```

```python
# Split data into random train and test subsets
train_data, test_data = train_test_split(basketball_df, test_size = 0.2, random_state=42)

X_train = train_data[['height', 'weight', 'success_field_goals', 'success_free_throws']].values
y_train = train_data['avg_points_scored'].values
X_test = test_data[['height', 'weight', 'success_field_goals', 'success_free_throws']].values
y_test = test_data['avg_points_scored'].values

print("X_train shape :",X_train.shape)
print("X_test shape  :",X_test.shape)
```

```python
# Standardize features by removing the mean and scaling to unit variance
scaler = StandardScaler().fit(X_train)

train_scaled = scaler.transform(X_train)
test_scaled = scaler.transform(X_test)
```

### statsmodels OLS

```python
# fit a linear regression line using statsmodels.
# augment the data with a column vector of 1's
train_data_augmented = sm.add_constant(train_scaled)
test_data_augmented = sm.add_constant(test_scaled)

OLSModel = OLS(y_train, train_data_augmented).fit()

print(OLSModel.summary())
```

```python
# R^2 regression score function on test data
y_pred = OLSModel.predict(test_data_augmented)
print("Statsmodel OLS R^2 Score on test data : {:.4f}".format(r2_score(y_test, y_pred)))
```

### sklearn Linear Regression

```python
# Linear Regression from sklearn
lr_model = LinearRegression()

# fit the model on train set
lr_model.fit(train_scaled, y_train)

print("Intercept : ", lr_model.intercept_)
print("Coefficients : ", lr_model.coef_)
```

```python
# make the prediction using fitted model and calculate R^2 regression score function on test data
y_pred = lr_model.predict(test_scaled)
print("Sklearn LinearRegression R^2 Score on test data : {:.4f}".format(r2_score(y_test, y_pred)))
```
