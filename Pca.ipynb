```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

import warnings
warnings.filterwarnings("ignore")

%matplotlib inline
%config InlineBackend.print_figure_kwargs = {'bbox_inches':None}
```

```python
# read the data
iris_df = pd.read_csv('data/iris.csv')
```

```python
# get the first 5 rows
iris_df.head()
```

```python
# Plot pairwise relationships in a dataset.
ax = sns.pairplot(iris_df, hue='Species')
```

```python
X = iris_df[['Sepal.Length', 'Sepal.Width', 'Petal.Length', 'Petal.Width']]
y = iris_df['Species'].astype('category').cat.codes
```

### eigenvalues and eigenvectors

```python
# Standardize features by removing the mean and scaling to unit variance
scaler = StandardScaler()

# Fit to data, then transform it.
X_scaled = scaler.fit_transform(X)
```

```python
# Compute the mean 
X_mean = np.mean(X_scaled, axis=0)

# Subtract the mean 
X = X_scaled - X_mean

# Estimate a covariance matrix
cov = np.cov(X.T)

print("covariance matrix \n", cov)
```

```python
# Compute the eigenvalues and eigenvectors of covaraince matrix
eigvals, eigvecs = np.linalg.eig(cov)

print("eigenvalues \n", eigvals)
print("\neigenvectors \n", eigvecs)
```

```python
# The proportion of the variation explained by a component is just 
# its eigenvalue divided by the sum of the eigenvalues

eigen_df = pd.DataFrame(eigvals, columns=['eigenvalues'])
eigen_df['explained variance %'] = eigen_df['eigenvalues'] / eigen_df['eigenvalues'].sum() * 100
eigen_df['cumulative explained varaince %'] = eigen_df['explained variance %'].cumsum()
eigen_df.index.name = 'principle component'

# display
eigen_df.round(3)
```

### Principal component analysis (PCA)

```python
# Fit the model with X_scaled and apply the dimensionality reduction on X_scaled.
X_reduced = PCA(n_components=3).fit_transform(X_scaled)

# plot the first three PCA dimensions
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(X_reduced[:,0], X_reduced[:,1], X_reduced[:,2], c=y, s=30, cmap=plt.cm.Set1)
ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([])
ax.set_xlabel('1st principle component', fontsize=14)
ax.set_ylabel('2nd principle component', fontsize=14)
ax.set_zlabel('3rd principle component', fontsize=14)
ax.set_title('PCA', fontsize=16)
```
